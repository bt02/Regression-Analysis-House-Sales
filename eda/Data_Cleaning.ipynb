```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('csv-files/kc_house_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(['view'], axis=1, inplace=True)
```

```python
for col in df.columns:
    print(col, '\n', df[col].value_counts(normalize=True).head(), '\n\n')
```

```python
df['waterfront'].fillna('?', inplace=True)

wf_prob = [0.0992434, 0.007566]
prob = [i/sum(wf_prob) for i in wf_prob]

def impute_waterfront(value):
    if value == '?':
        return np.random.choice([0.0, 1.0], p = prob)
    else:
        return value
    
df.waterfront = df.waterfront.map(lambda x: impute_waterfront(x))
```

```python
print('yr_renovated is missing', (round(df.yr_renovated.isna().sum() / len(df) * 100)), '% of total values')
```

```python
df.drop(['yr_renovated'], axis = 1, inplace=True)
```

```python
df['date'] = pd.to_datetime(df.date, format='%m/%d/%Y')
```

```python
df['sqft_basement'].value_counts(normalize=True)
```

```python
numbers=[]
for value in df['sqft_basement']:
    if value != '?':
          numbers.append(float(value))
```

```python
from statistics import median
df['sqft_basement'] = df['sqft_basement'].replace(['?'],median(numbers))
```

```python
df['sqft_basement'] = df['sqft_basement'].astype('float')
```

```python
df.set_index('id')
```

```python
df.to_csv('kc_house_data_clean.csv')
```
